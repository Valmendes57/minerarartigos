```python
texto = "O que é inteligência artificial (IA)?A inteligência artificial (IA) é um conjunto de tecnologias que permitem aos computadores executar uma variedade de funções avançadas, incluindo a capacidade de ver, entender e traduzir idiomas falados e escritos, analisar dados, fazer recomendações e muito mais. A IA é a espinha dorsal da inovação na computação moderna, agregando valor para indivíduos e empresas. Por exemplo, o reconhecimento óptico de caracteres (OCR) usa IA para extrair texto e dados de imagens e documentos, transformando conteúdo não estruturado em pronto para negócios, dados estruturados e insights valiosos.  Tudo pronto para começar? Clientes novos recebem US$ 300 em créditos para usar no Google Cloud."
palavras = word_tokenize(texto,language='portuguese')
print(palavras)
```

Frequência


```python
from nltk.probability import FreqDist
```

```python
fdisk_palavras = FreqDist(palavras)
```

```python
fdisk_palavras
```

```python
fdisk_palavras.most_common(10)
```

```python
fdisk_palavras.plot(10)
```

```python
!pip install wordcloud
from wordcloud import WordCloud , STOPWORDS
```

```python
def plot_cloud(wordcloud):
  plt.figure(figsize=(40,30))
  plt.imshow(wordcloud)
  plt.axis('off')
```

```python
wordcloud = WordCloud(width=3000, height=2000, random_state=1, background_color='black', colormap='Set2', collocations=False, stopwords=STOPWORDS).generate(texto)
```

```python
! pip install matplotlib
import matplotlib.pyplot as plt
```

```python
plot_cloud(wordcloud)
```
